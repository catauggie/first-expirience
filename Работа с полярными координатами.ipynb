```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('Книга1.xlsx')
```

```python
data
```

```python
a = data[['1_1','1_2', '1_3']].mean(axis=1)
b = data[['2_1','2_2', '2_3']].mean(axis=1)
c = data[['3_1','3_2', '3_3']].mean(axis=1)
```

```python
data.insert(5, "av1", np.round(a, 2))
data.insert(9, "av2", np.round(b, 2))
data.insert(13, "av3", np.round(c, 2))
```

```python
data
```

```python
df_sorted = data.sort_values(by=['Порядок'])
```

```python
plt.figure(figsize=(8, 8))
plt.axes(projection = 'polar')
  
# creating an array
rads = np.linspace(0, 23, 24)*15
  
# plotting the ellipse
dots=[]
for _ in range(len(rads)):
    s = np.random.randint(40,50)
    dots.append(s)
plt.polar(rads, dots, 'g.')#, linestyle='--')
    
plt.show()
```

```python
theta = np.arange(0, 2, 1./180)*np.pi
r = abs(4*np.sin(2*theta))
r2 = 2 + 0*theta
r3 = np.minimum(r, r2)
plt.polar(theta, r, lw=3)
plt.polar(theta, r2, lw=3)
plt.fill_between(theta, 0, r3, alpha=0.2)
plt.show()
```
