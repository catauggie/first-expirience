```python
import pandas as pd
import numpy as np
```

# Загрузка и обработка исходных данных

```python
data = pd.read_excel('Книга1.xlsx')
```

```python
data
```

```python
a = data[['1_1','1_2', '1_3']].mean(axis=1)
b = data[['2_1','2_2', '2_3']].mean(axis=1)
c = data[['3_1','3_2', '3_3']].mean(axis=1)
```

```python
data.insert(5, "av1", np.round(a, 2))
data.insert(9, "av2", np.round(b, 2))
data.insert(13, "av3", np.round(c, 2))
```

```python
data
```

```python
df_sorted = data.sort_values(by=['Порядок'])
```

```python
df_sorted.reset_index(drop= True , inplace= True )
```

```python
df_sorted
```

# Просто графики с осями

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.axes(projection = 'polar')

 
# creating an array
# containing the radian values
radians = list((df_sorted['Порядок']-1)* 2 *np.pi/24)
azims = list(df_sorted['av1'])

radians.append(radians[0])
azims.append(azims[0])
plt.polar(radians, azims, 'g.', linestyle='-')
    
theta = np.arange(0, 2, 1./180)*np.pi
plt.fill_between(theta, 80, 100, alpha=0.2, color='#FF1493')
plt.fill_between(theta, 60, 80, alpha=0.2, color='#00FF00')
plt.fill_between(theta, 40, 60, alpha=0.2, color='#008000')
plt.fill_between(theta, 20, 40, alpha=0.2, color='#FFD700')
    # display the polar plot
plt.show()
```

# Все три графика вместе

```python
plt.figure(figsize=(8, 8))
plt.axes(projection = 'polar')
  
# creating an array
# containing the radian values
radians = list((df_sorted['Порядок']-1)* 2 *np.pi/24)
azims1 = list(df_sorted['av1'])
azims2 = list(df_sorted['av2'])
azims3 = list(df_sorted['av3'])

radians.append(radians[0])
azims1.append(azims1[0])
azims2.append(azims2[0])
azims3.append(azims3[0])

plt.polar(radians, azims1, 'g.', linestyle='-', label='1 серия')
plt.polar(radians, azims2, 'c.', linestyle='-', label='2 серия')
plt.polar(radians, azims3, 'r.', linestyle='-', label='3 серия')
    
theta = np.arange(0, 2, 1./180)*np.pi
plt.fill_between(theta, 80, 100, alpha=0.2, color='#FF1493')
plt.fill_between(theta, 60, 80, alpha=0.2, color='#FFD700')
plt.fill_between(theta, 40, 60, alpha=0.2, color='#008000')
plt.fill_between(theta, 20, 40, alpha=0.2, color='#00FF00')
    # display the polar plot
plt.legend()
plt.show()
```
